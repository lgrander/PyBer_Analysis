```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
stdev = pyber_ride_df.std()
stdev["Avg. Fare ($USD)"]
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', color="skyblue")
pr = pyber_ride_df.groupby("Month")
# mean = pr.mean()
# errors = pr.std()
plt.errorbar(x="Month", y="Avg. Fare ($USD)", yerr=stdev, capsize=3, data=pyber_ride_df, color="black")
plt.yticks(np.arange(0, 60, 5))
plt.show()
```
